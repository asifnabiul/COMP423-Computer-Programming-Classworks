## Password checker

```python
def is_valid_length(possible_password):
    if len(possible_password) >= 8 and len(possible_password) <= 12:
        return True
    else:
        return False
```

```python
passwd = input('Enter new password: ')

if is_valid_length(passwd):
    print('Length is OK!')
else:
    print('Password length not acceptable!')
```
